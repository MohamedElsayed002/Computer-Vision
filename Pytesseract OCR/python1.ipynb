```python
import pytesseract
import cv2 
import matplotlib.pyplot as plt
import os 
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Teseract-OCR\tesseract.exe'

img = cv2.imread('./py.png')
```

```python
hImg , wImg , _ = img.shape 

boxes = pytesseract.image_to_boxes(img,lang='eng')

print(len(boxes))
```

```python
for b in boxes.splitlines():
    print(b)
    b = b.split(' ')
    print(b)
    x,y,w,h = int(b[1]),int(b[2]),int(b[3]),int(b[4])
    # cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
    cv2.rectangle(img,(x,hImg-y),(w,hImg-h),(0,255,0),3)
    cv2.putText(img, b[0], (x, hImg-y+50), cv2.FONT_HERSHEY_COMPLEX, 1, (255,0 ,0), 3)


plt.imshow(img)
```

```python
# Let's see how we will detecting wors 

boxes = pytesseract.image_to_data(img,lang='eng')

for x , b in enumerate(boxes.splitlines()):
    if x != 0:
        b = b.split()
        print(b)
        if len(b) == 12:
            x , y , w ,h = int(b[6]),int(b[7]),int(b[8]),int(b[9])
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
            cv2.putText(img, b[11], (x, y-5), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 3)

plt.imshow(img)
```
