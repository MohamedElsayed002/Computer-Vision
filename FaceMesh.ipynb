```python
import mediapipe as mp
import cv2
```

```python
image = cv2.imread('./mo.jpg')
```

```python
# Face Mesh 

# Load the image
rgb_image = cv2.imread('./mo.jpg')
# Convert to RGB if it's in BGR format
rgb_image = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB)
```

```python
print(rgb_image.shape)  # Should print something like (height, width, 3)
```

```python
if rgb_image is None:
    raise ValueError("Image not loaded. Check the path and file.")
```

```python
mp_face_mesh = mp.solutions.face_mesh.FaceMesh()
```

```python
result = mp_face_mesh.process(rgb_image)
```

```python
height , width , _ = image.shape 

for facial_landmarks in result.multi_face_landmarks:
    for i in range(0,468):
        pt1 = facial_landmarks.landmark[i]
        print(pt1)
        x = int(pt1.x * width)
        y = int(pt1.y * height)
        cv2.circle(image, (x,y), 2 , (100,100,0),-1)

cv2.imshow('Image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
