```python
import cv2 
import time 
import os 
```

```python
# Choose you camera number

camera = cv2.VideoCapture(0)
starting_time = time.time()
frame_counter = 0 
cap_frame = 0
dir_name = 'caputure_images'
number_image_captured = 20 
capture_image = False 

while True:

    # Checking the directory exist or not 
    isExist = os.path.exists(dir_name)

    if not isExist:
        os.makedirs(dir_name)

    frame_counter += 1
    
    # reading the frames from camera 
    ret , frame = camera.read()
    # reading saving frame seperatly 
    ret , saving_frame = camera.read()
    # finding with and height of image 
    height , width , dim = saving_frame.shape 

    # Writing the width and height of on the image saving image 
    cv2.putText(saving_frame,f'Height: {height}',(30,50),cv2.FONT_HERSHEY_COMPLEX,0.4,(0,255,0),1)
    cv2.putText(saving_frame,f'Width: {width}',(30,80),cv2.FONT_HERSHEY_COMPLEX,0.4,(0,255,0),1)

    # Capture frame if caputer_image is true and number_image_captured 
    if capture_image == True and cap_frame <= number_image_captured:
        cap_frame+=1 
        cv2.putText(frame,'Capturing',(50,70),cv2.FONT_HERSHEY_COMPLEX,0.4,(255,0,0),1)
        cv2.imwrite(f'{dir_name}/frame-{cap_frame}.png',saving_frame)
    else:
        cv2.putText(frame,'Press C to capture',(50,70),cv2.FONT_HERSHEY_COMPLEX,0.4,(255,0,0),1)
        cap_frame = 0
        capture_image = False 
    
    cv2.imshow('frame',frame)
    cv2.imshow('saving frame',saving_frame)
    total_time = time.time()
    frame_time = total_time - starting_time

    # Calculating how much frame pass in each second 
    fps = frame_counter / frame_time
    print(fps,'fps')
    if cv2.waitKey(1) == ord('q'):
        break
    if cv2.waitKey(1) == ord('c'):
        capture_image = True    

camera.release()
cv2.destroyAllWindows()

```
